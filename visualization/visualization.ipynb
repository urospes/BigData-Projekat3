```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#filepath = "/home/uros/Master/BigData/Projekti/Projekat3/results/predictions.csv"
filepath = "../results/predictions.csv"
data = pd.read_csv(filepath, sep=";", names=["event_time", "timestep", "time_period", "id", "x", "y", "cluster_pred", "fuel", "speed", "type", "noise", "emission_pred"])
data.head(5)
```

```python
data = gpd.GeoDataFrame(data, geometry=gpd.points_from_xy(data["x"], data["y"]))
data.head(5)
```

```python
world = gpd.read_file('./athens.geojson')
```

```python
world.head(5)
```

```python
world.crs
```

```python
clusters = data["cluster_pred"].unique()
clusters
```

```python
colors = ["red", "blue", "green", "yellow", "black", "white", "orange", "pink", "magenta"]
```

```python
world = world[world["name"] == "Athinaion"]
m = world[["geometry", "name"]].explore(figsize=(15, 15), fill=True, fillOpacity=0.1, highlight=False, tooltip=False)
for cluster in clusters:
    grp = data[data["cluster_pred"] == cluster]
    print(type(grp))
    grp.explore(m=m, color=colors[cluster], tooltip="emission_pred", marker_kwds=dict(radius=5, fill=True))

m
```
